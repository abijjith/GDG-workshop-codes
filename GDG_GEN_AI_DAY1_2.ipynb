```python
import nltk
from gensim.models import Word2Vec
nltk.download('punkt')

text = "Natural language processing is a fascinating field of artificial intelligence. It involves the interaction between computers and humans through natural language."

sentences = nltk.sent_tokenize(text)  # Split text into sentences
tokenized_sentences = [nltk.word_tokenize(sentence.lower()) for sentence in sentences]  # Tokenize each sentence into words

model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

unique_words = set(word for sentence in tokenized_sentences for word in sentence)

for word in unique_words:
    similar_words = model.wv.most_similar(word, topn=5)
    print(f"Most similar words to '{word}':")
    for similar_word, similarity in similar_words:
        print(f"- {similar_word} (similarity: {similarity:.4f})")
    print()

model.save('word2vec.model')

# Optional: Load the saved model
# loaded_model = Word2Vec.load('word2vec.model')
```
